```python
import sympy as sp


A = sp.Symbol('A')
c0 = sp.Symbol('c0')
wn = sp.Symbol('wn')
t = sp.Symbol('t')
x_0 = sp.Symbol('x_0')
f1 = x_0 + A*(1-sp.exp(-c0*t))*sp.sin(wn*t)
f1_diff = sp.diff(f1, t)
f1_d_diff = sp.diff(f1_diff, t)
f1_d_diff_simplified = sp.simplify(f1_d_diff)
sp.pprint(f1_d_diff_simplified)
```
